#Anova or F Test

ANOVA (Analysis of Variance) is used to determine whether there are significant differences between the means of three or more independent groups. It helps to avoid multiple comparisons, which would increase the risk of Type I error (false positives). It is particularly useful in experimental design, clinical trials, business analytics, and social sciences.

 **Types of ANOVA**





1.   One-Way ANOVA – Compares means between multiple groups based on one
independent variable.
2.   Two-Way ANOVA – Examines the effect of two independent variables simultaneously.
3.   Repeated Measures ANOVA – Used when the same subjects are tested multiple times under different conditions.



**Step-by-Step Example: One-Way ANOVA**

**Scenario**: Comparing Student Test Scores Based on Teaching Methods
A researcher wants to compare the effectiveness of three different teaching methods on student test scores. Three groups of students were taught using:

Method A
Method B
Method C

After the course, the students' test scores were recorded.

A	  B   C
85	78	88
90	82	91
88	80	86
92	84	85
87	79	90
The goal is to determine if at least one of the teaching methods leads to significantly different test scores.



**step1-Define Hypothesis**

Null hypothese:(**h0**)= There is no difference between 3 sample means.

Alternate hypothesis:(**h1**)=  At least one group has a different mean score among three, or there is a differenc between 3 sample means.

This is a **one tail** test which compares means between multiple groups based on one independent variable.



**Step2**- **Choose a significant level(alpha)**

choose a significant level(alpha)
the significance level is the proability of rejecting H0 when it is actually true
common choices:0.05(5%)or 0.01(1%),
we choose alpha=0.05

**Step3- Calculate Group Means and Overall Mean**

*   Sample Mean of Method A:  Xbar(A) =88.4
*   Sample Mean of Method B:  Xbar(B) =80.6
*   Sample Mean of Method C:  Xbar(C) =88.0


The overall Mean: Xbar= (88.4+80.6+88.0)/3 = 85.67

**Step4- Calculate the Anova formula (SSB)**

**Step 4- Compute Degrees of Freedom**

dfb between samples: k-1= 3-1= 2
dfw within smaples: N-k = 15-3= 12


**Step 5- Compute Mean Squares**

MSB= SSB/dfb= 193.45/2= 96.73

MSW= SSW/dfw= 79/12= 6.58

**Step 6- Compute the F-Statistic**

F= MSB/MSW= 96.73/6.58= 14.7

**Step 7: Compare with Critical Value**

Using an F-table at α=0.05 with dfb=2 and dfw=12

Critical F-value = 3.89.
Since 14.7 > 3.89, we reject the null hypothesis.

**Step 8: Conclusion**
Since the F-statistic is greater than the critical value, we conclude that at least one teaching method significantly affects test scores.
