```python
ZSCORE=HOW FAR A DATAPOINT IS FROM THE MEAN IN SD
VALUES BEYOND +-3 ARE CONSIDERED AS OUTLIERS
```

```python
import numpy as np
import pandas as pd
data={'value':[10,12,14,15,100,16,17,18,20]}
df=pd.DataFrame(data)
df
```

```python
df['z_score']=(df['value']-df['value'].mean())/df['value'].std()
df['z_score']
```

```python
mean_value=df['value'].mean()
std_dev=df['value'].std()
```
